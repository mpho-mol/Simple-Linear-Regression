```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Load the "insurance.csv" file into a pandas DataFrame
df = pd.read_csv("insurance.csv")
df.head()
```

```python
#Extract the relevant columns for age and charges
age = df['age']
charges = df['charges']
```

```python
# Plot a scatter plot of age against charges
plt.scatter(age, charges)
plt.xlabel("Age")
plt.ylabel("Charges")
plt.title("Insurance Costs vs Age")
plt.show()
```

```python
# Fit a linear regression model and make predictions
regression = LinearRegression()
X = age.values.reshape(-1, 1)
y = charges.values.reshape(-1, 1)
regression.fit(X, y)
predictions = regression.predict(X)
```

```python
# Plot another scatter plot with the best-fit line
plt.scatter(age, charges)
plt.plot(age, predictions, color='red')
plt.xlabel("Age")
plt.ylabel("Charges")
plt.title("Insurance Costs vs Age")
plt.show()
```
